```python
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()

for i in ['round','venue','formation','referee','opponent','team']:
    matk[i] = label.fit_transform(matk[i])

matk.head()
```

```python
matk.dtypes
```

```python
matk.isnull().sum()
```

```python
train = matk[matk["date"] < '2022-01-01']

test = matk[matk["date"] > '2022-01-01']
```

```python
train.head()
```

```python
import lazypredict
from lazypredict.Supervised import LazyClassifier
clf = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)
```

```python
train.columns
```

```python
predictors = ['round', 'venue', 'gf', 'ga', 'opponent', 'xg', 'xga', 'poss',
       'attendance', 'formation', 'referee', 'sh', 'sot', 'dist', 'fk', 'pk',
       'pkatt', 'season', 'team']
```

```python
train[predictors]
```

```python
test.head()
```

```python
models,predictions = clf.fit(train[predictors],test[predictors], train['target'], test['target'])
```

```python
print(models)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ada = AdaBoostClassifier(n_estimators=100, random_state=0)
ada.fit(train[predictors],train['target'])
```

```python
preds = ada.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
error = accuracy_score(test["target"], preds)
error
```
